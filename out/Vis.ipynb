```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("2threads.csv")
```

```python
size = data[u'size']
tseq = data[u' tseq']
tpar = data[u' tpar']
```

```python
dif = []
for _ in range(len(tseq)):
    dif.append(tseq[_]-tpar[_])
```

```python
plt.plot(size, dif, color='orange')
plt.ylabel('Difference')
plt.xlabel('Size')
plt.title('2 Threads Difference')
plt.show()
```

```python
plt.plot(size[0:100-1], tseq[0:100-1], color='red')
plt.plot(size[0:100-1], tpar[0:100-1], color='blue')
plt.ylabel('Time')
plt.xlabel('Red: Seq, Blue: Par')
plt.title('4 Threads')
plt.show()
```
